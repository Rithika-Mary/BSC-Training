```python
#1 What is Natural Language Processing (NLP)? Explain in your own words how it acts as a bridge between human language and computers.
Natural Language Processing (NLP) is a branch of Artificial Intelligence that helps computer understand and process human language.
It acts as a bridge by converting human language into a format that computers can understand and respond to.
```

```python
#2 Why do computers need NLP to understand human language? Give one real-world example to support your answer.
Computers Understand numbers, not human language. NLP helps convert text into a form that computers can process.
A Common Eg is Google Translate, Which translates one language into another.
```

```python
#3 List any four real-life applications of NLP and briefly describe what each one does.
Chatbots : Answer customer questions automatically.
Google Translate : Translates text from one language to another.
Spam Detection : Identifies unwanted emails.
Sentiment Analysis : Finds whether a review is positive or negative.
```

```python
#4 What is a rule-based NLP system? Explain using the spam detection example.
A rule-based NLP system works using predefined rules.
For eg, if a message contains the word "Congratulations", it can be marked as Possible Spam based on that rule.
```

```python
#5 List and briefly explain the seven steps of the NLP pipeline in the correct order.
1.Text Collection : Collect text data.
2.Text Cleaning : Remove unwanted symbols and punctuation.
3.Tokenization : Split text into words.
4.Stop Word Removal : Remove common words like "is" and "the".
5.Stemming/Lemmatization : Convert words to their root form.
6.Feature Extraction : Convert text into numbers.
7.Model Training : Train the machine learning model.
```

```python
#6 What is tokenization? Why is it a necessary step before a machine learning model can process text?
Tokenization is the process of splitting text into smaller parts like words or sentences.
It is necessary because machine learning models work with tokens instead of full sentences.
```

```python
#7 What are stopwords? Give three examples of stopwords in English and explain why they are usually removed.
Stopwords are common words that usually do not add much meaning to a sentence.
They are often removed during text processing.
Eg: is,the,and.
```

```python
#8 Why is it important to convert all text data to lowercase before processing it in NLP? Illustrate with an example.
Converting text to lowercase helps avoid treating the same word as different words.
Eg: "Phone" and "phone" become "phone", so they are treated as the same word.
```

```python
#9 List any four common mistakes beginners make while working with pandas for NLP tasks, as covered in the module.
1.Forgetting to import pandas.
2.Using the wrong column name.
3.Using = instead of == while filtering.
4.Forgetting to handle missing values.
```

```python
#10.Explain the difference between using = and == while filtering rows in a DataFrame, and why mixing them up causes errors.
= is used to assign a value to a variable, while == is used to compare values.
Using = instead of == in filtering causes an error because Python expects a comparison, not an assignment.
```

```python
#P6
import pandas as pd
data = {
    "Review": [
        "Good phone",
        "Bad battery",
        "Excellent camera",
        "Poor quality"
    ],
    "Sentiment": [
        "Positive",
        "Negative",
        "Positive",
        "Negative"
    ]
}
df = pd.DataFrame(data)
print(df)
```

```python
#P9
print(df["Sentiment"].value_counts())
```

```python
#P10
df["Review"] = df["Review"].str.lower()
print(df)
```

```python
#P11
df["Length"] = df["Review"].apply(len)
print(df)
```
