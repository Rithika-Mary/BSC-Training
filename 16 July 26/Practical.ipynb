```python
#25
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

sentences = [
    "I deposited money in the bank.",
    "The children played near the bank of the river."
]

for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    index = inputs.tokens().index("bank")
    embedding = outputs.last_hidden_state[0, index]

    print(sentence)
    print(embedding[:10])
    print()
```

```python
#26
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
model = SentenceTransformer("all-MiniLM-L6-v2")
sentences = [
    "Artificial Intelligence",
    "Machine Learning",
    "I love pizza"
]
embeddings = model.encode(sentences)
similarity = cosine_similarity(embeddings)
print(similarity)
```

```python
#27
import torch
import math

X = torch.rand(3, 4)
Q = X
K = X
V = X

scores = torch.matmul(Q, K.T)
scores = scores / math.sqrt(Q.shape[1])
weights = torch.softmax(scores, dim=-1)
context = torch.matmul(weights, V)

print("Attention Weights")
print(weights)
print("\nContext Vector")
print(context)
```

```python
#29
import torch
scores = torch.rand(5, 5)
mask = torch.triu(torch.ones(5, 5), diagonal=1)
scores = scores.masked_fill(mask == 1, float("-inf"))
attention = torch.softmax(scores, dim=-1)

print("Masked Scores")
print(scores)
print("\nAfter Softmax")
print(attention)
```

```python
#30
import torch
import torch.nn as nn

class DecoderBlock(nn.Module):
    def __init__(self):
        super().__init__()
        self.attention = nn.MultiheadAttention(32, 4, batch_first=True)
        self.norm1 = nn.LayerNorm(32)
        self.ffn = nn.Sequential(
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 32)
        )
        self.norm2 = nn.LayerNorm(32)

    def forward(self, x):
        attn, _ = self.attention(x, x, x)
        x = self.norm1(x + attn)
        ffn = self.ffn(x)
        x = self.norm2(x + ffn)
        return x

model = DecoderBlock()
x = torch.rand(2, 5, 32)
output = model(x)
print(output.shape)
```
