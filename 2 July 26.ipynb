```python
#Q40. Perform an A/B Test on Monthly Watch Time: Group A = Work_Mode = Office,Group B = Work_Mode = Hybrid. Calculate the mean watch time for both groups,perform an Independent T-Test, interpret significance at a = 0.05, and conclude whether the Hybrid work model results in significantly higher watch time
group_a = df[df["Work_Mode"] == "Office"]["Monthly_Watch_Time"]
group_b = df[df["Work_Mode"] == "Hybrid"]["Monthly_Watch_Time"]
print("Office Mean Watch Time:", group_a.mean())
print("Hybrid Mean Watch Time:", group_b.mean())
t_stat, p_value = ttest_ind(group_a, group_b)
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

#The A/B test compares the average Monthly Watch Time of Office and Hybrid employees. If the p-value is less than 0.05, the difference is statistically significant. Otherwise, the observed difference could simply be due to random variation, and there is not enough evidence to conclude that the Hybrid work model results in higher watch time.
```

```python
#Part H – Advanced Statistical Challenges
from sklearn.linear_model import LinearRegression
from scipy import stats
import numpy as np
```

```python
#Q41. Calculate the 95% confidence interval for the population mean Salary and interpret what it tells the company about pay estimates.
salary = df["Salary"]
mean_salary = salary.mean()
standard_error = stats.sem(salary)
confidence_interval = stats.t.interval(
    confidence=0.95,
    df=len(salary)-1,
    loc=mean_salary,
    scale=standard_error
)
print("Mean Salary:", mean_salary)
print("95% Confidence Interval:", confidence_interval)

#The 95% confidence interval gives the range within which the true average employee salary is likely to lie. It helps management estimate average salary with confidence.
```

```python
#Q42. Run an independent-samples t-test comparing Performance_Rating between the AI and Finance departments. State the hypotheses and conclude at a = 0.05.
ai = df[df["Department"] == "AI"]["Performance_Rating"]
finance = df[df["Department"] == "Finance"]["Performance_Rating"]
t_stat, p_value = stats.ttest_ind(ai, finance)

print("T Statistic:", t_stat)
print("P Value:", p_value) 

alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")

#This test checks whether employees in the AI and Finance departments have significantly different performance ratings.
```

```python
#Q43. Compute Z-Scores for Salary and Identify Outliers
df["Salary_Zscore"] = stats.zscore(df["Salary"])
outliers = df[abs(df["Salary_Zscore"]) > 2]
print(outliers[["Employee_ID", "Salary", "Salary_Zscore"]])

#Outlier salaries may represent senior management, highly experienced employees, or salaries that require further review.
```

```python
#Q44. Chi-Square Test (Department vs Left Company)
table = pd.crosstab(df["Department"], df["Left_Company"])
print(table)
chi2, p_value, dof, expected = stats.chi2_contingency(table)
print("Chi Square:", chi2)
print("P Value:", p_value)
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")

#This analysis determines whether employee attrition is related to department. If significant, management may focus retention efforts on specific departments.
```

```python
#Q45. One-Way ANOVA (Monthly Sales across Work Modes)
office = df[df["Work_Mode"] == "Office"]["Monthly_Sales"]
hybrid = df[df["Work_Mode"] == "Hybrid"]["Monthly_Sales"]
remote = df[df["Work_Mode"] == "Remote"]["Monthly_Sales"]
f_stat, p_value = stats.f_oneway(office, hybrid, remote)
print("F Statistic:", f_stat)
print("P Value:", p_value) 
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis")
else:
    print("Fail to Reject the Null Hypothesis")

#ANOVA compares the average monthly sales across Office, Hybrid, and Remote work modes. If the result is significant, work mode may influence employee sales performance.
```

```python
#Q47.Calculate Cohen's d (Office vs Hybrid Monthly Watch Time)
# Office group
office = df[df["Work_Mode"] == "Office"]["Monthly_Watch_Time"]

# Hybrid group
hybrid = df[df["Work_Mode"] == "Hybrid"]["Monthly_Watch_Time"]

# Means
mean1 = office.mean()
mean2 = hybrid.mean()

# Standard deviations
std1 = office.std()
std2 = hybrid.std()

# Sample sizes
n1 = len(office)
n2 = len(hybrid)

# Pooled Standard Deviation
pooled_std = np.sqrt(((n1-1)*(std1**2) + (n2-1)*(std2**2)) / (n1+n2-2))

# Cohen's d
cohen_d = (mean2 - mean1) / pooled_std

print("Cohen's d:", cohen_d)

#Cohen's d measures the practical difference in Monthly Watch Time between Office and Hybrid employees. A larger value indicates a stronger effect of the work mode.
```

```python
#Q48. Bootstrapping (95% Confidence Interval for Median Training Hours)
# Training Hours
training = df["Training_Hours"]

# Store bootstrap medians
bootstrap_medians = []

# Bootstrap sampling
for i in range(1000):
    sample = training.sample(frac=1, replace=True)
    bootstrap_medians.append(sample.median())

# Confidence Interval
lower = np.percentile(bootstrap_medians, 2.5)
upper = np.percentile(bootstrap_medians, 97.5)

print("95% Confidence Interval:")
print(lower, upper)

#This confidence interval estimates the likely range of the median training hours, helping management understand employee training patterns.
```

```python
#Q49. Detect Outliers in Monthly Sales Using IQR Method
# First Quartile
Q1 = df["Monthly_Sales"].quantile(0.25)

# Third Quartile
Q3 = df["Monthly_Sales"].quantile(0.75)

# IQR
IQR = Q3 - Q1

# Limits
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Outliers
outliers = df[(df["Monthly_Sales"] < lower_limit) |
              (df["Monthly_Sales"] > upper_limit)]

print("Outliers:")
print(outliers)

# Remove outliers
clean_data = df[(df["Monthly_Sales"] >= lower_limit) &
                (df["Monthly_Sales"] <= upper_limit)]

print("Mean Before:", df["Monthly_Sales"].mean())
print("Mean After :", clean_data["Monthly_Sales"].mean())

#The IQR method identifies unusually high or low sales values. Comparing the mean before and after removing outliers shows how extreme values influence the average sales.
```

```python
#Q50. Bayes' Theorem 
#P(Left Company | Performance Rating ≤ 3 AND Fraud Score > 0.5)
condition = df[(df["Performance_Rating"] <= 3) &
               (df["Fraud_Score"] > 0.5)]

# Employees satisfying the condition
condition_count = len(condition)

# Employees satisfying condition and left
left_condition = len(condition[condition["Left_Company"] == "Yes"])

# Posterior Probability
posterior_probability = left_condition / condition_count

print("Posterior Probability:", posterior_probability)

#A higher posterior probability suggests that employees with lower performance ratings and higher fraud scores are more likely to leave the company. This information can help identify high-risk employees and support proactive retention and risk management strategies.
```

```python
#Skewness
skewness = df["Salary"].skew()

print("Skewness:", skewness)

if skewness > 0:
    print("The salary distribution is Positively Skewed.")
elif skewness < 0:
    print("The salary distribution is Negatively Skewed.")
else:
    print("The salary distribution is Symmetrical.")

#The skewness value shows whether the salary distribution is balanced or tilted. A positive skew indicates more employees have lower salaries with a few employees earning much higher salaries.
```

```python
#Kurtosis
kurtosis = df["Salary"].kurt()

print("Kurtosis:", kurtosis)

if kurtosis > 0:
    print("The distribution is Leptokurtic.")
elif kurtosis < 0:
    print("The distribution is Platykurtic.")
else:
    print("The distribution is Mesokurtic.")

#Kurtosis measures the shape of the salary distribution. It helps identify whether the salary data contains unusually extreme values compared to a normal distribution.
```
