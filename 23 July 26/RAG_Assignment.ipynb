```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
text = """
Artificial Intelligence (AI) enables machines to perform tasks that normally require human intelligence.

Machine Learning is a branch of AI where systems learn from data.

Deep Learning uses neural networks with multiple layers.

Retrieval-Augmented Generation (RAG) combines document retrieval with Large Language Models.

RAG reduces hallucinations by retrieving relevant information before generating responses.

Embeddings convert text into numerical vectors.

FAISS is a library used for fast similarity search.

LangChain helps developers build RAG applications.

Chunking divides large documents into smaller pieces.

Cosine similarity measures the similarity between two vectors.
"""

with open("knowledge.txt", "w") as file:
    file.write(text)

print("Knowledge file created successfully.")
```

```python
with open("knowledge.txt", "r") as file:
    document = file.read()

print(document)
```

```python
# Split the document into smaller chunks

chunks = [chunk.strip() for chunk in document.split("\n\n") if chunk.strip()]

print("Number of Chunks :", len(chunks))

for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}:")
    print(chunk)
```

```python
# Load the Sentence Transformer model

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

print("Embedding model loaded successfully.")
```

```python
# Convert chunks into embeddings

embeddings = embedding_model.encode(chunks)

print("Embedding Shape :", embeddings.shape)
```

```python
# Create a FAISS index

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings, dtype="float32"))

print("Total vectors stored :", index.ntotal)
```

```python
# Function to retrieve the most relevant chunk

def retrieve_context(query):

    # Convert query into embedding
    query_embedding = embedding_model.encode([query])

    # Search in FAISS
    distance, index_number = index.search(
        np.array(query_embedding, dtype="float32"),
        k=1
    )

    return chunks[index_number[0][0]]
```

```python
question = "What is RAG?"

context = retrieve_context(question)

print("Question:")
print(question)

print("\nRetrieved Context:")
print(context)
```

```python
questions = [

    "What is Artificial Intelligence?",

    "What is Machine Learning?",

    "What is RAG?",

    "What is FAISS?",

    "Why is chunking used?"

]

for question in questions:

    print("="*60)

    print("Question:")
    print(question)

    context = retrieve_context(question)

    print("\nRetrieved Context:")
    print(context)

    print()
```

```python
query = "What is RAG?"

query_embedding = embedding_model.encode([query])

distance, index_number = index.search(
    np.array(query_embedding, dtype="float32"),
    k=3
)

print("Query :", query)

print("\nTop 3 Matching Chunks\n")

for i in range(3):

    print(f"Rank {i+1}")

    print("Chunk :")
    print(chunks[index_number[0][i]])

    print("Distance :", distance[0][i])

    print("-"*60)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
documents = [

    "Artificial Intelligence enables machines to perform intelligent tasks.",

    "Machine Learning allows computers to learn from data.",

    "Deep Learning uses neural networks.",

    "RAG combines retrieval with large language models.",

    "Embeddings convert text into vectors.",

    "FAISS performs fast similarity search.",

    "LangChain helps build LLM applications.",

    "Chunking divides long documents into smaller parts.",

    "Cosine similarity measures semantic similarity.",

    "Vector databases store embeddings."
]

print("Total Documents :", len(documents))
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

print("Model Loaded Successfully")
```

```python
embeddings = model.encode(documents)

print("Embedding Shape :", embeddings.shape)
```

```python
embeddings = np.array(embeddings).astype("float32")

faiss.normalize_L2(embeddings)

print("Embeddings Normalized")
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatIP(dimension)

index.add(embeddings)

print("Vectors Stored :", index.ntotal)
```

```python
def semantic_search(query, top_k=3):

    query_embedding = model.encode([query])

    query_embedding = np.array(query_embedding).astype("float32")

    faiss.normalize_L2(query_embedding)

    scores, indices = index.search(query_embedding, top_k)

    print("\nQuery :", query)

    print("\nTop Results\n")

    for score, idx in zip(scores[0], indices[0]):

        print("Document :", documents[idx])

        print("Similarity :", round(float(score),4))

        print("-"*50)
```

```python
semantic_search("What is RAG?")
```

```python
semantic_search("What are embeddings?")

semantic_search("Explain FAISS")

semantic_search("What is Deep Learning?")
```

```python
query = input("Enter your question : ")

semantic_search(query)
```

```python
print("Semantic Search Completed Successfully")
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

from langchain_huggingface import HuggingFaceEmbeddings

from langchain_community.vectorstores import FAISS
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
pdf_name = list(uploaded.keys())[0]

loader = PyPDFLoader(pdf_name)

documents = loader.load()

print("Pages Loaded :", len(documents))
```

```python
splitter = RecursiveCharacterTextSplitter(

    chunk_size=500,

    chunk_overlap=100

)

chunks = splitter.split_documents(documents)

print("Chunks Created :", len(chunks))
```

```python
embedding_model = HuggingFaceEmbeddings(

    model_name="sentence-transformers/all-MiniLM-L6-v2"

)

print("Embedding Model Ready")
```

```python
vector_store = FAISS.from_documents(

    chunks,

    embedding_model

)

print("Vector Store Created")
```

```python
retriever = vector_store.as_retriever(

    search_kwargs={"k":3}

)

print("Retriever Ready")
```

```python
question = input("Enter your question : ")

results = retriever.get_relevant_documents(question)
```

```python
for i, doc in enumerate(results):

    print("="*60)

    print("Result", i+1)

    print(doc.page_content)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss
```

```python
documents = [
    {"topic":"AI","text":"Artificial Intelligence enables machines to think."},
    {"topic":"ML","text":"Machine Learning learns from data."},
    {"topic":"DL","text":"Deep Learning uses neural networks."},
    {"topic":"RAG","text":"RAG combines retrieval with LLMs."},
    {"topic":"FAISS","text":"FAISS performs fast similarity search."},
    {"topic":"Embedding","text":"Embeddings convert text into vectors."}
]

texts = [doc["text"] for doc in documents]
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(texts)

embeddings = np.array(embeddings).astype("float32")
```

```python
dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(embeddings)

print("Vectors Stored:", index.ntotal)
```

```python
def search(query):

    query_embedding = model.encode([query])

    query_embedding = np.array(query_embedding).astype("float32")

    distance, indices = index.search(query_embedding,3)

    return distance, indices
```

```python
query = "Explain Retrieval Augmented Generation"

distance, indices = search(query)

print("Optimized Search Results\n")

for idx in indices[0]:

    print(documents[idx]["topic"])

    print(documents[idx]["text"])

    print()
```

```python
topic = "RAG"

print("Metadata Filtering\n")

for doc in documents:

    if doc["topic"] == topic:

        print(doc["text"])
```

```python
query = "What is AI?"

distance, indices = search(query)

print("Top 3 Documents\n")

for idx in indices[0]:

    print(texts[idx])
```

```python
#RAG Evaluation
retrieved_documents = [
    "Artificial Intelligence enables machines to think.",
    "Machine Learning learns from data.",
    "Deep Learning uses neural networks.",
    "RAG combines retrieval with LLMs.",
    "FAISS performs similarity search."
]

print("Retrieved Documents Loaded")
```

```python
relevant_documents = [
    "RAG combines retrieval with LLMs.",
    "FAISS performs similarity search."
]

print("Relevant Documents Loaded")
```

```python
def precision_at_k(retrieved, relevant, k):

    retrieved_k = retrieved[:k]

    relevant_found = 0

    for doc in retrieved_k:

        if doc in relevant:

            relevant_found += 1

    precision = relevant_found / k

    return precision
```

```python
k = 3

precision = precision_at_k(
    retrieved_documents,
    relevant_documents,
    k
)

print("Precision@", k, "=", round(precision,2))
```

```python
def recall_at_k(retrieved, relevant, k):

    retrieved_k = retrieved[:k]

    relevant_found = 0

    for doc in retrieved_k:

        if doc in relevant:

            relevant_found += 1

    recall = relevant_found / len(relevant)

    return recall
```

```python
recall = recall_at_k(
    retrieved_documents,
    relevant_documents,
    5
)

print("Recall =", round(recall,2))
```

```python
print("Retrieved Documents\n")

for doc in retrieved_documents:

    print(doc)
```

```python
print("Relevant Documents\n")

for doc in relevant_documents:

    print(doc)
```

```python
print("Evaluation Results")

print("------------------------")

print("Precision@3 :", round(precision,2))

print("Recall@5 :", round(recall,2))
```
