```python
# Encapsulation - Bank Account

class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def check_balance(self):
        print("Balance:", self.__balance)

account = BankAccount()

account.deposit(5000)
account.withdraw(1000)
account.check_balance()
```

```python
# Polymorphism

class SpamDetector:
    def predict(self):
        print("Spam Prediction Completed")

class FaceRecognition:
    def predict(self):
        print("Face Detected")

class ImageClassifier:
    def predict(self):
        print("Image Classified")

models = [SpamDetector(), FaceRecognition(), ImageClassifier()]

for model in models:
    model.predict()
```

```python
# Abstraction

from abc import ABC, abstractmethod

class Model(ABC):

    @abstractmethod
    def train(self):
        pass

class SpamDetector(Model):
    def train(self):
        print("Training Spam Detector")

class FaceRecognition(Model):
    def train(self):
        print("Training Face Recognition Model")

m1 = SpamDetector()
m2 = FaceRecognition()

m1.train()
m2.train()
```

```python
# AI Spam Detection System

import logging

logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

class SpamDetector:

    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    def predict(self, message):

        if message.strip() == "":
            raise ValueError("Message cannot be empty")

        spam_words = ["prize", "congratulations", "winner", "offer"]

        for word in spam_words:
            if word.lower() in message.lower():
                return "Spam"

        return "Not Spam"


detector = SpamDetector("Spam Shield", "1.0")

try:

    message = input("Enter Message: ")

    result = detector.predict(message)

    print("Prediction:", result)

    file = open("prediction.txt", "a")
    file.write(result + "\n")
    file.close()

    confidence = "94%"

    logging.info(f"Prediction: {result}")
    logging.info(f"Confidence: {confidence}")

except ValueError as e:
    print("Error:", e)
```

```python
# AI Spam Detection System

import logging

logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

class SpamDetector:

    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    def predict(self, message):

        if message.strip() == "":
            raise ValueError("Message cannot be empty")

        spam_keywords = ["congratulations", "prize", "winner", "offer"]

        for word in spam_keywords:
            if word in message.lower():
                return "Spam"

        return "Not Spam"


try:

    detector = SpamDetector("Spam Shield", "1.0")

    message = input("Enter Message: ")

    result = detector.predict(message)

    print("Prediction:", result)

    with open("prediction.txt", "a") as file:
        file.write(result + "\n")

    confidence = "94%"

    logging.info(f"Prediction: {result}")
    logging.info(f"Confidence: {confidence}")

    print("Prediction saved to prediction.txt")
    print("Log saved to ai_system.log")

except ValueError as e:
    print("Error:", e)
```

```python
#OOP + File Handling
class Student:

    def __init__(self, name, course):
        self.name = name
        self.course = course

    def save(self):

        file = open("students.txt", "a")

        file.write(f"{self.name}, {self.course}\n")

        file.close()


s1 = Student("Rithika", "AI & ML")

s1.save()

print("Student data saved successfully.")

#Without file handling, data is stored only in memory and is lost when the program ends.
#By saving the data into a file (students.txt), it remains available even after closing the program.
```
