```python
#Understand the role of logging.basicConfig()
import logging

logging.basicConfig(
    filename="app.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
```

```python
import logging

logging.basicConfig(
    filename="gps_log.txt",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

latitude = 12.9716
longitude = 77.5946

logging.info(f"GPS Location: Latitude={latitude}, Longitude={longitude}")

print("Location logged successfully.")
```
