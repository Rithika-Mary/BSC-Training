```python
# Checking for missing values in each column
df.isnull().sum()
```

```python
# Fill missing numerical values with median
# Fill missing categorical values with mode

if df.isnull().sum().sum()==0:
    print("No missing values found.")
else:
    df.fillna(df.median(numeric_only=True), inplace=True)

    for col in df.select_dtypes(include='object'):
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
## Check for duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate records from the dataset
df.drop_duplicates(inplace=True)

print("Duplicate records removed successfully.")
```

```python
# Visualize outliers in numerical columns using a box plot
plt.figure(figsize=(8,5))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.xticks(rotation=90)
plt.show()
```

```python
#Univariate Analysis
# Plot the distribution of employee ages
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Plot the distribution of monthly salaries
plt.figure(figsize=(6,4))
sns.histplot(df["Monthly_Salary"], kde=True)
plt.title("Monthly Salary Distribution")
plt.show()
```

```python
# Display the number of employees in each department
plt.figure(figsize=(6,4))
sns.countplot(x="Department", data=df)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a pivot table showing average salary by department and gender
pivot = pd.pivot_table(
    df,
    values="Monthly_Salary",
    index="Department",
    columns="Gender",
    aggfunc="mean"
)

pivot
```

```python
# Visualize the pivot table using a heatmap

plt.figure(figsize=(8,5))

sns.heatmap(
    pivot,
    annot=True,
    cmap="YlGnBu"
)

plt.title("Average Salary by Department and Gender")

plt.show()
```
